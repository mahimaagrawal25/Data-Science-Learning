```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head()
```

```python
columns=df.columns
columns
```

```python
titles = df.get('movie_title')
titles
```

```python
for t in titles:
    length = len(t)
    print(length)
```

```python
freq_title = {}

for t in titles:
    length = len(t)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1

print(freq_title)
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X, Y)
plt.xlabel("Movie title len")
plt.ylabel("No of Movies")
plt.show()
```
